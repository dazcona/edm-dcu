```python
import datetime

def get_dates(col, weeks=4):
    base = df[col].min()
    limit = df[col].max()
    dates = []
    x = base.replace(day=1) # first day of the month
    while x < limit:
        dates.append(x)
        x += datetime.timedelta(weeks=weeks)
    return dates
```

```python
indices = ['A', 'B', 'C', 'D', 'E', 'X', 'S']
```

```python
list(enumerate(indices))
```
